```python
#linear search
def lin(a,idx,key):
    if idx>=len(a):
        return False
    elif a[idx]==key:
        return True
    else:
        return lin(a,idx+1,key)
lin([15,8,13,20,17,1],0,30)
```

```python
#binary search
def bin(a,l,r,key):
    if l>r:
        return False
    else:
        mid=(l+r)//2
        if a[mid]==key:
            return True
        elif key<a[mid]:
            return bin(a,l,mid-1,key)
        else:
            return bin(a,mid+1,r,key)
print(bin([3,5,13,28,32,33,41,45,48],0,8,13))
```

```python
#30 
def groupSum5(start, nums, target): 
    # Base Case 
    if start >= len(nums): 
        return target == 0
  
    # If the current element is a multiple of 5 
    if nums[start] % 5 == 0: 
  
        # Check if the next element is 1 
        if start < len(nums)-1 and nums[start+1] == 1: 
            return groupSum5(start + 2, nums, target - nums[start]) 
        else: 
            return groupSum5(start + 1, nums, target - nums[start]) 
  
    # else try both the cases that include and exclude the current element 
    return groupSum5(start + 1, nums, target - nums[start]) or groupSum5(start + 1, nums, target)


groupSum5(0, [2, 5, 10, 4], 22)
```

```python
def countRotations(arr, n):

    min = arr[0]
    min_index = 0
    for i in range(0, n):

        if (min > arr[i]):
            min = arr[i]
            min_index = i

    return min_index;





arr = [20, 30, 40, 50, 60, 10]
N = len(arr)
print(countRotations(arr, N))
```

```python
def countRotations(arr, n):
    start = 0
    end = n - 1


    while start <= end:
        mid = start + (end - start) // 2


        prev = (mid - 1 + n) % n
        nex = (mid + 1) % n


        if arr[mid] < arr[prev] \
                and arr[mid] <= arr[nex]:
            return mid


        elif arr[mid] < arr[start]:
            end = mid - 1
        elif arr[mid] > arr[end]:
            start = mid + 1
        else:
            return 0



arr = [20, 30, 40, 50, 60, 10]
N = len(arr)
print(countRotations(arr, N))
```
