```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
class DoublyList:
  
  def __init__(self, a):
  #  Design the constructor based on data type of a. If 'a' is built in python list then
  #  Creates a linked list using the values from the given array.
    if type(a)==list:
        self.head = Node(a[0],None,None)
        temp=self.head
        for i in range(1,len(a)):
            newNode=Node(a[i],None,temp)
            temp.next=newNode
            temp=temp.next
            
    else:
        self.head=a
    


  
  # Counts the number of Nodes in the list
  def countNode(self):
    count=0
    temp=self.head.next
    while (temp!=None):
        count+=1
        temp=temp.next
    return count


  
  # prints the elements in the list
  def forwardprint(self):
    # To Do
    temp=self.head.next        #10node
    while (temp.next!=None):    #10,20,30
        print(temp.element,end=",")
        temp=temp.next         #10node=>20node
    print(str(temp.element)+".")

  # prints the elements in the list backward
  def backwardprint(self):
    # To Do
    temp=self.head.prev
    while (temp.prev!=None):
        print(temp.element,end=",")    
        temp=temp.prev
    print(str(temp.element)+".")

  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    # To Do
    count=self.countNode()
    if (idx>=count or idx<0):
        return None
    else:
        temp=self.head.next
        count=0
        while(temp!=None):
            if count==idx:
                return temp
            count+=1
            temp=temp.next
    return Node("Index Error",None,None)
  
  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    # To Do
    count=0
    n=self.head.next
    while(n!=None):
      if(n.element==elem):
        return count
      count+=1
      n=n.next
    return -1

  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):
    # To Do
    count=self.countNode()
    if(idx>count or idx<0):
      return None
    elif(idx==count):
      n=self.nodeAt(idx-1)
      newNode=Node(elem,None,None)
      n.next=newNode
      newNode.prev=n
      newNode.next=self.head
      self.head.prev=newNode
    else:
      n=self.nodeAt(idx)
      pred=n.prev
      newNode=Node(elem,None,None)
      pred.next=newNode
      newNode.next=n
      newNode.prev=n.prev
      n.prev=newNode
  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
  def remove(self, idx):
    # To Do
    count=self.countNode()
    if(idx>=count or idx<0):
      return None
    else:
      n=self.nodeAt(idx)
      pred=n.prev
      next=n.next
      pred.next=next
      next.prev=pred
      n.next=None
      n.prev=None
      return str(n.element)

    
```
