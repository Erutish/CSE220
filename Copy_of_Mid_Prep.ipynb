```python
#Linear Array
#Question1

a = [6,2,None]
b = [85,25,1,3,2,54,6,None,None]

n = [None]*(len(a)+len(b))

p = 0
for i in b:
  if i not in n:
    n[p] = i
    p = p+1

p = 5 #size of b
for i in a:
  if i not in n:
    n[p] = i
    p = p+1  

c = 0
i = 0
while(n[i]!=None):
  c += 1
  i += 1
print(c)  
```

```python
#Circular Array
#Question 1
a = [1,2,1]
b = 3
c = [None]*b

i = 0
m = 0
max = 0
while(i<len(a)):
  j = 0
  max = -1
  while(j<len(a)):
    if a[j]>a[i]:
      max = a[j]
    j = j+1  
  c[m] = max 
  i = i+1
  m = m+1
print(c)  
```

```python
#Question2
a =  [ 1, 6, 4, 3, 5 ]
b = 5
c = [None]*b

i = 0
m = 0
max = 0
while(i<len(a)):
  max = -1
  j = i
  if i == len(a)-1:
    while(j<len(a)):
      if a[j]>a[i]:
        max = a[j]
        break
      j = (j+1)%len(a)
  else:  
    while(j<len(a)):
      if a[j]>a[i]:
        max = a[j]
        
        break
      j = j+1  
  c[m] = max
  i += 1
  m += 1
print(c)      
```
