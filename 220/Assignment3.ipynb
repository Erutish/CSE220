```python
class Node:
  def __init__(self, e, n):
    self.element = e
    self.next = n

class LinkedList:
  
  def __init__(self, a):
  #  Design the constructor based on data type of a. If 'a' is built in python list then
  #  Creates a linked list using the values from the given array. head will refer
  #  to the Node that contains the element from a[0]
  #  Else Sets the value of head. head will refer
  # to the given LinkedList
  # Hint: Use the type() function to determine the data type of a
    if type(a)==list:
      self.head=Node(a[0],None)
      temp=self.head
      for i in range(1,len(a)):
          newNode=Node(a[i],None)
          temp.next=newNode
          temp=temp.next
    else:
      self.head=a


  
  # Count the number of nodes in the list
  def countNode(self):
    temp=self.head
    count=0
    while (temp!=None):
        count+=1
        temp=temp.next
    return count
  
  # Print elements in the list
  def printList(self):
    # To Do
    temp=self.head
    while temp.next!=None:
        print (temp.element,end=",")
        temp=temp.next
    else:
        print(temp.element)

  # returns the reference of the Node at the given index. For invalid index return None.
  def nodeAt(self, idx):
    # To Do
    length=self.countNode()
    temp=self.head
    if idx>=length:
        return None
    else:
        for i in range(idx):
            temp=temp.next
    return temp
        


  
  # returns the element of the Node at the given index. For invalid idx return None.
  def get(self, idx):
    # To Do
    length=self.countNode()
    temp=self.head
    if idx>=length:
        return None
    else:
        for i in range(idx):
            temp=temp.next
    return temp.element

  
  # updates the element of the Node at the given index. 
  # Returns the old element that was replaced. For invalid index return None.
  # parameter: index, element
  def set(self, idx, elem):
    # To Do
    length=self.countNode()
    temp=self.head
    if idx>=length:
        return None
    else:
        #for i in range(idx):
            #emp=temp.next
        temp=self.nodeAt(idx)
    oldelement=temp.element
    temp.element=elem
    return oldelement


  # returns the index of the Node containing the given element.
  # if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    # To Do
    temp=self.head
    count=0
    while temp!=None:
        if temp.element==elem:
            return count
        temp=temp.next
        count+=1
    return "-1"

  
  # returns true if the element exists in the List, return false otherwise.
  def contains(self, elem):
    # To Do
    temp=self.head
    while temp!=None:
        if temp.element==elem:
            return True
        temp=temp.next
    return False

  # Makes a duplicate copy of the given List. Returns the reference of the duplicate list.
  def copyList(self):
    # To Do
    temp=self.head
    copy_head=Node(self.head.element,None)
    copy_temp=copy_head
    while temp.next!=None:
        newNode=Node(temp.next.element,None)
        temp=temp.next
        copy_temp.next=newNode
        copy_temp=copy_temp.next
    return copy_head


  # Makes a reversed copy of the given List. Returns the head reference of the reversed list.
  def reverseList(self):
    # To Do
    temp=self.head
    copy_head=Node(self.head.element,None)
    while temp.next!=None:
        newNode=Node(temp.next.element,None)
        newNode.next=copy_head
        copy_head=newNode
        temp=temp.next
    return copy_head
  
  # inserts Node containing the given element at the given index
  # Check validity of index.
  def insert(self, elem, idx):
    # To Do
    #temp=self.head
    newNode=Node(elem,None)
    if idx==0:
        newNode.next=self.head
        self.head=newNode
    else:
        prevNode=self.nodeAt(idx-1)
        newNode.next=prevNode.next
        prevNode.next=newNode
   

  # removes Node at the given index. returns element of the removed node.
  # Check validity of index. return None if index is invalid.
  def remove(self, idx):
    # To Do
    if idx==0:
        removeNode=self.head
        self.head=self.head.next
    else:
        prevNode=self.nodeAt(idx-1)
        removeNode=prevNode.next
        prevNode.next=removeNode.next
    
    return removeNode.element
  

  
  # Rotates the list to the left by 1 position.
  def rotateLeft(self):
    # To Do
    oldhead=self.head
    
    self.head=self.head.next
    temp=self.head
    while (temp.next!=None):
        temp=temp.next
    temp.next=oldhead
    oldhead.next=None
  
  
  # Rotates the list to the right by 1 position.
  def rotateRight(self):
    # To Do
    temp=self.head
    p=None
    while temp.next!=None:
        p=temp #last er ager elem
        temp=temp.next
    temp.next=self.head #just last and fist change korle e hoy cz baki sob linked
    self.head=temp
    p.next=None
        
        
```
