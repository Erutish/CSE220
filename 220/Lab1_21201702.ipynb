```python
#...................Task 1.................................

def shiftLeft(source,k):
    for j in range(k):
        for i in range(1,len(source)):
            source[i-1]=source[i]
        source[len(source)-1]=0
    print(source)
source=[10,20,30,40,50,60]
k=int(input('Enter a number:'))

shiftLeft(source,k)
```

```python
#....................Task 3....................

def shiftRight(source,k):
    for i in range(k):
        for i in range(len(source)-1,0,-1):
            source[i]=source[i-1]
        source[0]=0
    print(source)
source=[10,20,30,40,50,60]
k=int(input('Enter a number:'))
shiftRight(source,k)
```

```python
#............Task4....................

def rotateRight(source,k):
    for i in range(k):
        temp=source[len(source)-1]
        for j in range(len(source)-1,0,-1):
            source[j]=source[j-1]
        source[0]=temp
    print(source)
source=[10,20,30,40,50,60]
k=int(input('Enter a number:'))
rotateRight(source,k)
```

```python
#............................Task7............................

def equal(source):
    sum1=0
    sum2=0
    for i in range(1,len(source)):
        for m in range(0,i):
            sum1+=source[m]
        for n in range(i,len(source)):
            sum2+=source[n]
        if(sum1==sum2):
            return True
        sum1=0
        sum2=0
    return False

source= [1, 1, 1, 2, 1]
source1= [2, 1, 1, 2, 1] 
source2= [10, 3, 1, 2, 10]
print(equal(source))
print(equal(source1))
print(equal(source2))
```

```python
#/................Task8..............................
n=int(input("n="))
def arraySeries(n):
    arr=[None]*(n*n)
    k=0
    for i in range(1,n+1):
        for j in range(n,0,-1):
            if (j<=i):
                arr[k]=j
                k+=1
            else:
                arr[k]=0
                k+=1
    return arr
print(arraySeries(n))            


```
